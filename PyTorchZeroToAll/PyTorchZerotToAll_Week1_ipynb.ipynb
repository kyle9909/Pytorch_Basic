# Lec 02. Linear Model

## Model & Loss

```python
w = 1.0 # random value 설정

# Linear model
def forward(x):
  return x*w

# Loss function
def loss(x,y):
  y_pred = forward(x)
  return (y_pred-y) * (y_pred-y)
```

## Compute loss for w

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w_list = []
mse_list = []

for w in np.arange(0.0, 4.1, 0.1):
  print('w=', w)
  l_sum = 0
  for x_val, y_val in zip(x_data, y_data):
    y_pred_val = forward(x_val)
    l = loss(x_val, y_val)
    l_sum += l
    print('\t', x_val, y_val, y_pred_val, l)
  print('MSE=', l_sum/3)
  w_list.append(w)
  mse_list.append(l_sum/3)
```

## Visualize loss

```python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```

# Lec 03. Gradient Descent

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

# linear model
def forward(x):
  return x*w 

# loss function
def loss(x,y):
  y_pred = forward(x)
  return (y_pred-y)**2

# gradient
def gradient(x,y):
  return 2*x*(x*w-y)
```

```python
# before training
print("predict (before training)", 4, forward(4))


# train loop
for epoch in range(100):
  for x_val, y_val in zip(x_data, y_data):
    grad = gradient(x_val, y_val)
    w = w - 0.01*grad  # w 가중치 update
    print('\t grad:', x_val, y_val, grad)
    l = loss(x_val, y_val)

  print('progress:', epoch, 'w=', w, 'loss=', l)

# after training
print("predict (after training)", 4, forward(4))
```

## PyTorch Rhythm

### Step1. Design model using class and Variable

```python
import torch
from torch.autograd import Variable

# Data : 3*1 matrix
x_data = Variable(torch.Tensor([[1.0],[2.0],[3.0]]))
y_data = Variable(torch.Tensor([[2.0],[4.0],[6.0]]))

class Model(torch.nn.Module):
  def __init__(self):
    '''
    2개의 nn.Linear.module를 인스턴스화
    '''
    super(Model, self).__init__()
    # model 정의
    self.linear = torch.nn.Linear(1,1) #input&ouput 모두 1개

  def forward(self,x):
    y_pred = self.linear(x)
    return y_pred

# 사용할 model
model = Model()
```

### Step2. Construct loss and optimizer

```python
criterion = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
# model.parameters() : 어떤 parameter를 update 해야하는지!
```

### Step3. Training : forward, loss, backward, step

```python
# epoch에 따른 모델 학습
for epoch in range(500):
  # x를 통해 y_pred 계산
  y_pred = model(x_data)

  # loss 계산
  loss = criterion(y_pred, y_data)
  print(epoch, loss.data.detach().numpy())

  # Zero gradient, backwardpass, parameter updata
  optimizer.zero_grad() # 기존 gradient 초기화
  loss.backward()  # back propagation
  optimizer.step()  #Update
```

## Train CIFAR 10 Classifier

```python
from torch import nn
from torch import optim
from torch.nn import functional as F
import torchvision
import torchvision.transforms as transforms
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

batch_size=10
```

```python
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)
```

```python
# Step1. Design model using class
class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()
    self.conv1 = nn.Conv2d(3,6,5)
    self.pool = nn.MaxPool2d(2,2)
    self.conv2 = nn.Conv2d(6,16,5)
    self.fc1 = nn.Linear(16*5*5, 120)
    self.fc2 = nn.Linear(120,84)
    self.fc3 = nn.Linear(84,10)

  def forward(self,x):
    x = self.pool(F.relu(self.conv1(x)))
    x = self.pool(F.relu(self.conv2(x)))
    x = x.view(-1,16*5*5)
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.fc3(x)
    return x
  
net = Net()

# Step2. Define Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum = 0.9)

# Step3. Trian the network
for epoch in range(5):
  running_loss = 0
  for i, data in enumerate(trainloader,0):
    inputs, labels = data

    inputs, labels = Variable(inputs), Variable(labels)

    # 초기 graident 값 0
    optimizer.zero_grad()

    # Forward + Backward + optimize
    output = net(inputs)
    loss = criterion(output, labels)
    loss.backward()
    optimizer.step()

    # 학습 현황 출력
    running_loss += loss.data.detach().numpy()
    if i%2000 == 1999:  #2000 batch 마다 출력
      print('[%d, %5d] loss : %.3f'%
            (epoch + 1, i+1, running_loss/2000))
      running_loss = 0.0

print('Finish Training')
```

## MNIST Softmax

```python
import torchvision.datasets as datasets
```

```python
batch_size = 64
train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=True, download=True,
                   transform = transforms.Compose([
                                                   transforms.ToTensor(), transforms.Normalize((0.1307,),(0.3081,))])),
                                           batch_size=batch_size, shuffle=True)

test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=False,
                   transform = transforms.Compose([
                                                   transforms.ToTensor(), transforms.Normalize((0.1307,),(0.3081,))])),
                                           batch_size=batch_size, shuffle=True)

class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()
    self.l1 = nn.Linear(784,580)  #MNIST size : 1*28*28*28 (flatten)
    self.l2 = nn.Linear(580,320)
    self.l3 = nn.Linear(320,240)
    self.l4 = nn.Linear(240,120)
    self.l5 = nn.Linear(120,10)

  def forward(self,x):
    x = x.view(-1,784)  #flatten data
    x = F.relu(self.l1(x))
    x = F.relu(self.l2(x))
    x = F.relu(self.l3(x))
    x = F.relu(self.l4(x))
    return self.l5(x)   #logit 형태로 cross entropy 계산

model = Net()

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr = 0.01, momentum=0.5)

def train(epoch):
  model.train()
  for batch_idx, (data,target) in enumerate(train_loader):
    data, target = Variable(data), Variable(target)
    optimizer.zero_grad()
    output = model(data)
    loss = criterion(output,target)
    loss.backward()
    optimizer.step()
    if batch_idx % 100 == 0:
      print("Train epoch : {} [{}/{} ({:.0f}%)]\t Loss: {:.6f}".format(
          epoch, batch_idx*len(data), len(train_loader.dataset),
          100.*batch_idx/len(train_loader), loss.data.detach().numpy()))
```

```python
def test():
  model.eval()
  test_loss = 0
  correct = 0
  for data, target in test_loader:
    test_data, target = Variable(data, volatile=True), Variable(target)
    test_output = model(test_data)
    test_loss += criterion(test_output, target).data.detach().numpy()
    
    # 예측 값 중 최대 값을 출력
    pred = torch.max(test_output.data,1)[1]
    correct += pred.eq(target.data.view_as(pred)).cpu().sum()

  test_loss /= len(test_loader.dataset)  #loss의 평균
  print('\n Test set : Average loss : {:4f}, Accuracy : {}/{} ({:.0f}%) \n'.format(
      test_loss, correct, len(test_loader.dataset),
      100. *correct / len(test_loader.dataset)
  ))
```

```python
Y_pred_scores = Variable(torch.Tensor([[0.3, 0.2, 0.9, 0.1]]))

# torch.max ==> max value & indices를 모두 출력

Y_pred_val, Y_pred_idx = torch.max(Y_pred_scores.data,1)
print(Y_pred_val)
print(Y_pred_idx)
```

```python
for epoch in range(1,10):
  train(epoch)
  test()
```
